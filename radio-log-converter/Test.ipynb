```python
import os
import pandas as pd
from tabula import read_pdf
```

```python
def file_list(n_pages):
    return [f'PAGE {i}.pdf' for i in range(1, n_pages+1)]
```

```python
def to_dataframe(file_list):
    df_list = []
    for file in file_list:
        df = read_pdf(os.path.join(os.getcwd(), file), multiple_tables=True, pages='all')
        df_list.append(df)
    return df_list
```
